# Numerical Results for the A+B→C Network

### Compare Gillespie simulation with mass action kinetics and Plefka approximation

```julia
using Pkg
cd(@__DIR__)
Pkg.activate("../../")
using CRN

include("plotting.jl")
```

```julia
# Parameters
num_species = 3     # number of species
num_int = 1         # number of interaction reaction

# reaction constants
k1 = [8, 8, 3]
k2 = [3, 2, 1.5]
k3 = [0.5]
k = [k1, k2, k3]

# Stoichiometric
s_i = zeros(Int, num_int, num_species)
r_i = copy(s_i)
s_i[1,:] = [0 0 1]
r_i[1,:] = [1 1 0]

# Initial condition
x0 = k1 ./ k2

# times
t_init = 0.0
t_final = 5.0
delta_t = 0.01
tspan = collect(t_init:delta_t:t_final)

# Plefka Expansion parameter
α = 1.0
alphaOrder = 1
orderParameter = "linear"
plf_sim = Plefka(α, alphaOrder, orderParameter)
```

## Gillespie Simulation

```julia
N = 100000    # Repititions
res_gil = gillespie_avg(x0, k, s_i, r_i, tspan, N)
plot_trajectories(res_gil)
```

## Mass Action Kinetics

```julia
function f_massAction!(du,u,k,t)
    du[1] = k1[1] - k2[1]*u[1] - α*k3[1]*u[1]*u[2]
    du[2] = k1[2] - k2[2]*u[2] - α*k3[1]*u[1]*u[2]
    du[3] = k1[3] - k2[3]*u[3] + α*k3[1]*u[1]*u[2]
end

ts = (tspan[1], tspan[end])
res_ma = ode_solver(f_massAction!, k, ts, x0)
plot_trajectories_and_deviation(Result(tspan,res_ma.data'), res_gil)
```
